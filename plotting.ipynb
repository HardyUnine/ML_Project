### Defining plot function for price

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_price_from_csv(ticker):
    filename = f"{ticker}.csv"
    df = pd.read_csv(filename)
    df_sorted = df.sort_values(by='DAY')
    days = df_sorted['DAY']
    prices = df_sorted['PRICE_PER_SHARE']
    company_name = df_sorted['COMPANY_NAME'].iloc[0]
    
    plt.figure(figsize=(10,6))
    plt.plot(days, prices, marker='o', linestyle='-', color='b')
    plt.title(f"Price per Share Over Time for {company_name} [{ticker}]")
    plt.xlabel("Day")
    plt.ylabel("Price Per Share")
    plt.grid(True)
    plt.show()
```

```python
plot_price_from_csv("AAPL")  
plot_price_from_csv("AMC")
plot_price_from_csv("AMD")
```

```python
def plot_price_and_rsi(csv_path):
    """
    Plots Price Per Share and RSI on the same plot, using different y-axes.
    Left y-axis: Price per Share
    Right y-axis: RSI
    Args:
        csv_path (str): Path to the input CSV file
    """
    df = pd.read_csv(csv_path)
    ticker = df['TICKER'].iloc[0]
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax2 = ax1.twinx()
    # Price per share (bars or line)
    ax1.bar(df['DAY'], df['PRICE_PER_SHARE'], width=0.8, alpha=0.7, color='tab:blue', label='Price Per Share')
    ax1.set_ylabel('Price Per Share', color='tab:blue')
    ax1.set_xlabel('DAY')
    # RSI line
    ax2.plot(df['DAY'], df['RSI'], color='orange', linewidth=2, label='RSI')
    ax2.set_ylabel('RSI', color='orange')
    ax2.axhline(70, color='gray', linestyle='--', linewidth=0.8)  # Overbought
    ax2.axhline(30, color='gray', linestyle='--', linewidth=0.8)  # Oversold
    # Title and legends
    plt.title(f"{ticker} - Price Per Share & RSI Over Time")
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.tight_layout()
    plt.show()

plot_price_and_rsi("AAPL.csv")
plot_price_and_rsi("AMC.csv")
plot_price_and_rsi("AMD.csv")
```

```python
def plot_shorts_and_bf(csv_path):
    """
    Plots Shorts count and Borrow Fee on dual axes.
    Left y-axis: Number of Shorts
    Right y-axis: Borrow Fee
    Args:
        csv_path (str): Path to the input CSV file
    """
    df = pd.read_csv(csv_path)
    ticker = df['TICKER'].iloc[0]
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax2 = ax1.twinx()
    
    # Shorts on left axis
    ax1.plot(df['DAY'], df['SHORTS'], color='red', linewidth=2.5, label='Shorts', marker='o')
    ax1.set_ylabel('Number of Shorts', color='red', fontsize=11)
    ax1.set_xlabel('DAY')
    ax1.tick_params(axis='y', labelcolor='red')
    
    # Borrow Fee on right axis
    ax2.plot(df['DAY'], df['BF'], color='purple', linewidth=2.5, label='Borrow Fee', marker='s')
    ax2.set_ylabel('Borrow Fee (%)', color='purple', fontsize=11)
    ax2.tick_params(axis='y', labelcolor='purple')
    
    # Title and legends
    plt.title(f"{ticker} - Shorts & Borrow Fee Over Time")
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.tight_layout()
    plt.show()

plot_shorts_and_bf("AAPL.csv")
plot_shorts_and_bf("AMC.csv")
plot_shorts_and_bf("AMD.csv")
```
